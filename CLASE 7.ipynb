# Json Bajado

```python
import requests
import pprint
import json
```

```python
url ='https://cdn.buenosaires.gob.ar/datosabiertos/datasets/salud/hospitales/hospitales.geojson'

contenido = requests.get(url).text
#objeto = json.load(contenido) # Da error por q lo baja com strim
objeto = json.loads(contenido) #le agrego la s de string
pprint.pprint(objeto)
```

```python
objeto['features'][0]['properties']['NOMBRE']

#entro al primer nivel, al elemeto 0 a las propiedades y ahi elijo
```

```python
# nombre y direccion de los hospitales

for i in range(len(objeto['features'])):
    nombre =objeto['features'][i]['properties']['NOMBRE']
    direc =objeto['features'][i]['properties']['DOM_NORMA']
    print(nombre,'- ', direc)
```

# APIs

```python
direccion= 'Rivadavia y Loria'
localidad = 'caba'

#metodo de request para que parametrice 

dir_codoficada = requests.utils.quote(direccion+ ','+ localidad)
print(dir_codoficada)
```

```python
url = 'http://servicios.usig.buenosaires.gob.ar/normalizar/?geocodificar=True&direccion='+dir_codoficada
```

```python
url
```

```python
# Hacemos el request

respuesta = requests.get(url).text

#Uso json para parsear

dir_normalizar = json.loads(respuesta)
dir_normalizar
```

```python
print('Latitud: ',dir_normalizar['direccionesNormalizadas'][0]['coordenadas']['y'])
print('Longitud: ',dir_normalizar['direccionesNormalizadas'][0]['coordenadas']['x'])
```
