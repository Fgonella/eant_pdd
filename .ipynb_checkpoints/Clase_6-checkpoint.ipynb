```python
# imprtamos librerias

from bs4 import BeautifulSoup as BS
from selenium import webdriver
from time import sleep
```

```python
# declaramos la varibale y ebrimos el archivo de driver 

#Crome option para el navegador
option = webdriver.ChromeOptions()
#option.add_argument('headless') #no se abre el navegador y se lo agrego abajo 
#option.add_argument('--incognito') lo abre en modo incognito


driver = webdriver.Chrome(executable_path='chromedriver', options=option)

driver.get('https://www.olx.com.ar/items/q-aspiradoras-auto')

sleep(3)

#capturo el Boton
scrip_js = """
    let boton = document.querySelector('[data-aut-id="btnLoadMore"]') 
    if (boton){
            boton.click()
        }
    else {
            return "No existe"
    }
"""

while driver.execute_script(script_js) != "No existe":
    sleep(3)
    
sleep(2)

#Capturo el navegador para despues sacar la informacion

html = driver.execute_script('return document.documentElement.outerHTML')
#driver.quit() #Cierra el navegador


```

```python
#Ahora usamos libreria BS4 como ya sabemos 
dom = BS(html, features='html.parser')
```

```python
productos = dom.find_all(class_='IKo3_')

for i in range(len(productos)):
    desc = productos[i].find(class_='_2tW1I').string
    precio = productos[i].find(class_='_89yzn').string
    print(desc + ': ' + precio)
```

# Ahora Ejemplo con La nacion

* Los articulos aparecen cuando terminas de scrolear

```python
# declaramos la varibale y ebrimos el archivo de driver 

#Crome option para el navegador
option = webdriver.ChromeOptions()
#option.add_argument('headless') #no se abre el navegador y se lo agrego abajo 
#option.add_argument('--incognito') lo abre en modo incognito


driver = webdriver.Chrome(executable_path='chromedriver', options=option)

driver.get('https://www.lanacion.com.ar/')

sleep(3)

#Armo el script de java, java te da la posivilidad de retornar el valor 
scrip_scroll= """
    let fin_pantalla = document.body.scrollHeight
    window.scrollTo(0, fin_pantalla)
    return fin_pantalla
"""

pos_actual = 0 #es donde arranca a scrolear
pos_siguente = driver.execute_script(scrip_scroll)
sleep(3)

while pos_actual != pos_siguente:
    pos_actual = pos_siguente
    pos_siguente = driver.execute_script(scrip_scroll)
    sleep(3)
    print(pos_actual)
    
print('Legamos al final')

html = driver.execute_script('return document.documentElement.outerHTML')
driver.quit()
```

```python
dom = BS(html, features='html.parser')
```

```python
titulos = dom.find_all(class_="mod-description")

contador = 0

for i in range(len(titulos)):
    titulo = titulos[i].a['title']
    hiper = 'https://www.lanacion.com.ar'+titulos[i].a['href']
    contador +=1
    print (str(contador)+ ':'+ titulo + ', '+hiper+ '\n')
```
