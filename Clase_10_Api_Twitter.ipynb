```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret) 
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
usuario = api.me
```

```python
# Mongo Base de datos, quiero capturar los tweets y guardarlos 

bd = cliente['bigdata']
coleccion = bd['tweets']
```

```python
# Buscar el ultimo tweet, para no repetirlo en la base de datos (en este caso no tengo por q no guarde)

ultimo_tweet = coleccion.find_one(sort=list({'id':-1}.items()))
if ultimo_tweet !=None: 
    ultimo_id =ultimo_tweet['id']
else:
    ultimo_id = None
```

```python
#Since_ID -> Es desde el ultimo Id traeme todo 
#Yo quiero guardar los tweet

tweets=[]
contador = 1
for tweet in tweepy.Cursor(api.user_timeline,since_id=ultimo_id ,screen_name='fergonella' ,tweet_mode='extended').items(3000):
    tweet_dic = tweet._json
    tweets.append(tweet_dic)
    print('Tweets caputurado: ', contador)
    contador+=1
    
```

```python
#Ahora los subo a mongo si hay tweets 

if len(tweets)>0:
    coleccion.insert_many(tweets)
    print(f'subidos: {len(tweets)}')
else:
    printt('No hay para subir')
```
